```python
from faker import Faker
import random
import pandas as pd
from datetime import datetime
```

```python
# from faker import Faker
# import random

# Define the Zomato Data Generator class

class ZomatoDataGenerator:
    def __init__(self):
        self.fake = Faker()

    def create_customers(self, num_customers):
        customers = []
        for _ in range(num_customers):
            customers.append({
                "customer_id": _ + 1,
                "name": self.fake.name(),
                "email": self.fake.email(),
                "phone": self.fake.phone_number(),
                "address": self.fake.address(),
                "signup_date": self.fake.date_this_decade(),
                "is_premium": random.choice([True, False]),
                "preferred_cuisine": random.choice(['Indian', 'Chinese', 'Italian', 'Mexican', 'Japanese']),
                "total_orders": random.randint(1, 100),
                "average_rating": round(random.uniform(1, 5), 1)})
        return pd.DataFrame(customers)
    def create_restaurants(self, num_restaurants):
        cuisines = ["Indian", "Chinese", "Italian", "Mexican", "Japanese", "Thai"]
        restaurants = []
        for _ in range(num_restaurants):
            restaurants.append({
                "restaurant_id": _ + 1,
                "name": self.fake.company(),
                "cuisine_type": random.choice(['Indian', 'Chinese', 'Italian', 'Mexican', 'Japanese']),
                "location": self.fake.city(),
                "owner_name": self.fake.name(),
                "average_delivery_time": random.randint(20, 60),
                "contact_number": self.fake.phone_number(),
                "rating": round(random.uniform(1, 5), 1),
                "total_orders": random.randint(50, 100),
                "is_active": random.choice([True, False])})
        return pd.DataFrame(restaurants)
    def create_orders(self, num_orders):
        order_statuses = ["Pending", "Completed", "Cancelled"]
        orders = []
        for _ in range(num_orders):
            orders.append({
                "order_id": _ + 1,
                "customer_id": random.choice(['1001','2023', '9876']),
                "restaurant_id": random.choice(['1013', '5023', '7890']),
                "items": self.fake.sentence(nb_words=1),
                "quantity": random.randint(1, 5),
                "order_date": self.fake.date_this_year(),
                "delivery_time": self.fake.time(),
                "order_statuses": random.choice(['Pending', 'Delivered', 'Cancelled']),
                "total_amount": round(random.uniform(5.0, 100.0), 2),
                "payment_mode": random.choice(['Credit Card', 'Cash', 'UPI']),
                "discount_applied": round(random.uniform(0, 20), 2),
                "feedback_rating": round(random.uniform(1, 5), 1)})
        return pd.DataFrame(orders)
    def create_deliveries(self, num_deliveries):
        delivery_statuses = ["Delivered", "In Transit", "Failed"]
        deliveries = []
        for _ in range(num_deliveries):
            deliveries.append({
                "delivery_id": _ + 1,
                "order_id": random.choice(['1001', '2023', '9876']),
                "delivery_person_id": random.randint(1, 100),
                "delivery_status": random.choice(['On the way', 'Delivered']),
                "distance": round(random.uniform(1.0, 10.0), 2),
                "delivery_time": round(random.uniform(10, 60), 1),
                "estimated_time": random.randint(15, 30),
                "delivery_fee": round(random.uniform(2.0, 5.0), 2),
                "vehicle_type":random.choice(['Bike', 'Car', 'Scooter', 'Cycle'])})
        return pd.DataFrame(deliveries)
```

```python
mycursor.execute("USE ZOMATO_DB")
```

```python
mycursor.execute("""
    CREATE TABLE customers (
        customer_id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(255),
        phone VARCHAR(100),
        email VARCHAR(255),
        address VARCHAR(255),
        signup_date DATE,
        is_premium BOOLEAN,
        preferred_cuisine VARCHAR(100),
        total_orders INT,
        average_rating DECIMAL(5,1)
    )
""");
```

```python
mycursor.execute("""
    CREATE TABLE restaurants (
        restaurant_id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(255) NOT NULL,
        cuisine_type VARCHAR(100),
        location VARCHAR(255),
        owner_name VARCHAR(255),
        average_delivery_time INT, 
        contact_number VARCHAR(100),
        rating DECIMAL(2,1),       
        total_orders INT DEFAULT 0,
        is_active BOOLEAN DEFAULT TRUE
    )
""");
```

```python
mycursor.execute("""
    CREATE TABLE orders (
        order_id INT AUTO_INCREMENT PRIMARY KEY,
        customer_id INT,
        restaurant_id INT,
        items VARCHAR(255),
        quantity INT,
        order_date DATETIME,
        delivery_time TIME,
        order_statuses VARCHAR(100),
        total_amount DECIMAL(10,2),
        payment_mode VARCHAR(50),
        discount_applied DECIMAL(10,2),
        feedback_rating DECIMAL(2,1)
    )
""");
```

```python
mycursor.execute("""
    CREATE TABLE deliveries (
        delivery_id INT AUTO_INCREMENT PRIMARY KEY,
        order_id INT NOT NULL,
        delivery_person_id INT,
        delivery_status VARCHAR(50),
        distance DECIMAL(5,2),             
        delivery_time INT,                 
        estimated_time INT,                
        delivery_fee DECIMAL(6,2),
        vehicle_type VARCHAR(50)        
    )
""");
```
